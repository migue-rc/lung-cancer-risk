```python
import pandas as pd
df_lung = pd.read_csv('data/lung_cancer.csv')
df_lung
```

```python
print(df_lung.info())
df_lung.describe()
```

# In this notebook we are going to have two approach.
1. We are going to explore the data, in a unsupervised way.
2. We aregoing to expore the data in a supervised way.

```python

from sklearn.model_selection import train_test_split

target_label = 'lung_cancer_risk'
y = df_lung[target_label]
X = df_lung.drop(target_label, axis=1)
```

```python

from statsmodels_utils import plot_correlation_heatmap
plot_correlation_heatmap(X, 'Correlation Heatmap all features')
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
smoking_corr_features = ['smoking_years', 'cigarettes_per_day', 'pack_years']
df_smoker = df_lung[smoking_corr_features]
g = sns.PairGrid(df_smoker)
g.map_lower(sns.scatterplot, alpha=0.5, edgecolor='k')
g.map_diag(sns.histplot, fill=True, color='gray')
g.map_upper(lambda x, y, **kwargs: plt.gca().annotate(
    f"{x.corr(y):.2f}",
    xy=(0.5, 0.5),
    xycoords='axes fraction',
    ha='center',
    va='center',
    fontsize=16,
    fontweight='bold'
))
g.add_legend()
plt.show()

```

```python
seed = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed, stratify=y)

plt.hist(y_train)
plt.show()
```

There are some high correlations, possibly we are having a VIH situation.

```python
from statsmodels_utils import calculate_and_drop_vif

X_train_clean, dropped_features = calculate_and_drop_vif(X_train)
plot_correlation_heatmap(X_train_clean, 'Cleaned Correlation Heatmap')
```

```python
# go for stats model logit
from statsmodels_utils import stepwise_selection
import statsmodels.api as sm


logit_model = sm.Logit
final_model, features = stepwise_selection(logit_model, X_train_clean, y_train, disp=False)
print(f"Optimal features: {features}")
final_model.summary()
```

```python
from statsmodels_utils import plot_coef
plot_coef(final_model)
```

```python
from statsmodels_utils import plot_shap_summary
X_test_shap = X_test[features]

with_constant = sm.add_constant(X_test_shap)
plot_shap_summary(final_model, with_constant, plot_type='violin')
```

```python
from statsmodels_utils import plot_shap_waterfall_single
plot_shap_waterfall_single(final_model, with_constant, row_index=8)
```

```python
from statsmodels_utils import plot_shap_bar
plot_shap_bar(final_model, with_constant)
```

```python

from statsmodels_utils import plot_shap_dependence
plot_shap_dependence(final_model,with_constant, 'age')
```

```python
from statsmodels_utils import plot_top_shap_dependence
plot_top_shap_dependence(final_model, with_constant, top_n=3)
```
